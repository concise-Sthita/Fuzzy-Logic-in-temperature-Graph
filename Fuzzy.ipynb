```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x_qual = np.arange(0, 11, 1)
x_serv = np.arange(0, 11, 1)
x_tip = np.arange(0, 26, 1)

# Generate fuzzy membership functions
qual_lo = fuzz.trimf(x_qual, [0, 0, 5])
qual_md = fuzz.trimf(x_qual, [0, 5, 10])
qual_hi = fuzz.trimf(x_qual, [5, 10, 10])

serv_lo = fuzz.trimf(x_serv, [0, 0, 5])
serv_md = fuzz.trimf(x_serv, [0, 5, 10])
serv_hi = fuzz.trimf(x_serv, [5, 10, 10])

tip_lo = fuzz.trimf(x_tip, [0, 0, 13])
tip_md = fuzz.trimf(x_tip, [0, 13, 25])
tip_hi = fuzz.trimf(x_tip, [13, 25, 25])

# Visualize membership functions
plt.figure()

plt.plot(x_qual, qual_lo, 'b', linewidth=1.5, label='Poor')
plt.plot(x_qual, qual_md, 'g', linewidth=1.5, label='Average')
plt.plot(x_qual, qual_hi, 'r', linewidth=1.5, label='Good')
plt.title('Quality')
plt.legend()

plt.figure()

plt.plot(x_serv, serv_lo, 'b', linewidth=1.5, label='Poor')
plt.plot(x_serv, serv_md, 'g', linewidth=1.5, label='Average')
plt.plot(x_serv, serv_hi, 'r', linewidth=1.5, label='Good')
plt.title('Service')
plt.legend()

plt.figure()

plt.plot(x_tip, tip_lo, 'b', linewidth=1.5, label='Low')
plt.plot(x_tip, tip_md, 'g', linewidth=1.5, label='Medium')
plt.plot(x_tip, tip_hi, 'r', linewidth=1.5, label='High')
plt.title('Tip')
plt.legend()

plt.show()
```
