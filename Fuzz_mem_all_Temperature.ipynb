```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x_ot = np.arange(0, 35, 1)
x_rt = np.arange(0, 50, 1)

# Generate fuzzy membership functions
# Triangular membership functions
rt_lo_tri = fuzz.trimf(x_rt, [0, 0, 15])
rt_md_tri = fuzz.trimf(x_rt, [0, 15, 27])
rt_hi_tri = fuzz.trimf(x_rt, [15, 27, 35])

ot_lo_tri = fuzz.trimf(x_ot, [0, 0, 15])
ot_md_tri = fuzz.trimf(x_ot, [0, 15, 35])
ot_hi_tri = fuzz.trimf(x_ot, [15, 35, 50])

# Gaussian membership functions
rt_lo_gauss = fuzz.gaussmf(x_rt, 2, 1)
rt_md_gauss = fuzz.gaussmf(x_rt, 5, 1)
rt_hi_gauss = fuzz.gaussmf(x_rt, 8, 1)

ot_lo_gauss = fuzz.gaussmf(x_ot, 2, 1)
ot_md_gauss = fuzz.gaussmf(x_ot, 5, 1)
ot_hi_gauss = fuzz.gaussmf(x_ot, 8, 1)

# Trapezoidal membership functions
rt_lo_trap = fuzz.trapmf(x_rt, [0, 0, 2, 4])
rt_md_trap = fuzz.trapmf(x_rt, [2, 4, 6, 8])
rt_hi_trap = fuzz.trapmf(x_rt, [6, 8, 10, 10])

ot_lo_trap = fuzz.trapmf(x_ot, [0, 0, 2, 4])
ot_md_trap = fuzz.trapmf(x_ot, [2, 4, 6, 8])
ot_hi_trap = fuzz.trapmf(x_ot, [6, 8, 10, 10])


# Visualize membership functions
plt.figure(figsize=(10, 8))

# Triangular membership functions
plt.subplot(3, 1, 1)
plt.plot(x_rt, rt_lo_tri, 'b', linewidth=1.5, label='Low')
plt.plot(x_rt, rt_md_tri, 'g', linewidth=1.5, label='Average')
plt.plot(x_rt, rt_hi_tri, 'r', linewidth=1.5, label='High')
plt.title('Room Temperature - Triangular')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x_ot, ot_lo_tri, 'b', linewidth=1.5, label='Low')
plt.plot(x_ot, ot_md_tri, 'g', linewidth=1.5, label='Average')
plt.plot(x_ot, ot_hi_tri, 'r', linewidth=1.5, label='High')
plt.title('Outside Temperature - Triangular')
plt.legend()

plt.tight_layout()

# Gaussian membership functions
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(x_rt, rt_lo_gauss, 'b', linewidth=1.5, label='Poor')
plt.plot(x_rt, rt_md_gauss, 'g', linewidth=1.5, label='Average')
plt.plot(x_rt, rt_hi_gauss, 'r', linewidth=1.5, label='Good')
plt.title('Room Temperature - Gaussian')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x_ot, ot_lo_gauss, 'b', linewidth=1.5, label='Poor')
plt.plot(x_ot, ot_md_gauss, 'g', linewidth=1.5, label='Average')
plt.plot(x_ot, ot_hi_gauss, 'r', linewidth=1.5, label='Good')
plt.title('Outside Temperature - Gaussian')
plt.legend()

plt.tight_layout()

# Trapezoidal membership functions
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(x_rt, rt_lo_trap, 'b', linewidth=1.5, label='Low')
plt.plot(x_rt, rt_md_trap, 'g', linewidth=1.5, label='Average')
plt.plot(x_rt, rt_hi_trap, 'r', linewidth=1.5, label='High')
plt.title('Room Temperature - Trapezoidal')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x_ot, ot_lo_trap, 'b', linewidth=1.5, label='Low')
plt.plot(x_ot, ot_md_trap, 'g', linewidth=1.5, label='Average')
plt.plot(x_ot, ot_hi_trap, 'r', linewidth=1.5, label='High')
plt.title('Outside Temperature - Trapezoidal')
plt.legend()

plt.tight_layout()

plt.show()
```
