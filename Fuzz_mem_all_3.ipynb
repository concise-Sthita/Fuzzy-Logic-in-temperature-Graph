```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe variables
x_qual = np.arange(0, 11, 1)
x_serv = np.arange(0, 11, 1)
x_tip = np.arange(0, 26, 1)

# Generate fuzzy membership functions
# Trapezoidal membership functions
qual_lo_trap = fuzz.trapmf(x_qual, [0, 0, 2, 4])
qual_md_trap = fuzz.trapmf(x_qual, [2, 4, 6, 8])
qual_hi_trap = fuzz.trapmf(x_qual, [6, 8, 10, 10])

serv_lo_trap = fuzz.trapmf(x_serv, [0, 0, 2, 4])
serv_md_trap = fuzz.trapmf(x_serv, [2, 4, 6, 8])
serv_hi_trap = fuzz.trapmf(x_serv, [6, 8, 10, 10])

tip_lo_trap = fuzz.trapmf(x_tip, [0, 0, 5, 10])
tip_md_trap = fuzz.trapmf(x_tip, [5, 10, 15, 20])
tip_hi_trap = fuzz.trapmf(x_tip, [15, 20, 25, 25])

# Visualize membership functions
plt.figure(figsize=(10, 8))
# Trapezoidal membership functions
plt.figure(figsize=(10, 8))

plt.subplot(3, 1, 1)
plt.plot(x_qual, qual_lo_trap, 'b', linewidth=1.5, label='Poor')
plt.plot(x_qual, qual_md_trap, 'g', linewidth=1.5, label='Average')
plt.plot(x_qual, qual_hi_trap, 'r', linewidth=1.5, label='Good')
plt.title('Quality - Trapezoidal')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(x_serv, serv_lo_trap, 'b', linewidth=1.5, label='Poor')
plt.plot(x_serv, serv_md_trap, 'g', linewidth=1.5, label='Average')
plt.plot(x_serv, serv_hi_trap, 'r', linewidth=1.5, label='Good')
plt.title('Service - Trapezoidal')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x_tip, tip_lo_trap, 'b', linewidth=1.5, label='Low')
plt.plot(x_tip, tip_md_trap, 'g', linewidth=1.5, label='Medium')
plt.plot(x_tip, tip_hi_trap, 'r', linewidth=1.5, label='High')
plt.title('Tip - Trapezoidal')
plt.legend()

plt.tight_layout()

plt.show()
```
